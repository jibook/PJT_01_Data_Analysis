## 파일 합치기 (for구문, concat 사용)

- for구문

```python
hos_data = pd.DataFrame([])

f_list = ['./data/H_data/T20_2019_1백만.1.csv','./data/H_data/T20_2019_1백만.2.csv','./data/H_data/T20_2019_1백만.3.csv']

for fname in f_list:
    temp = pd.read_csv(fname, encoding='cp949')
    hos_data = hos_data.append(temp)

hos_data.head()
```

```python
hos_data.info()
```

- concat

```python
data_2019 = pd.concat([data_2019_01, data_2019_02, data_2019_03])
```

```python
data_2019.info()
```

# 연령대 별로 자주 걸리는 질병은?

## 빈도 수 확인
   - value.counts() 활용

```python
data_2019.value_counts()   # 2019_01 전체, 이것보단 컬럼별로 확인 필요
```

```python
# 연령에 상관 없이 가장 많이 진단받은 질병은?
data_2019['주상병코드'].value_counts()
```

```python
# 병원을 많이 찾은(=건보공단 보장을 많이 이용한) 연령은?
data_2019['연령대코드'].value_counts()
```

### Insight

**병원 이용 순위**

   - 2위 55~59세(12)

   - 5위 50~54세(11)
   - ...
   - 18위 15~19세(4)
   
 1위인 60 ~ 64세(13)와 마지막 순위인 15 ~ 19세(4)는 약 4배 차이

```python
# 연령대코드를 index로 설정(세팅, set), 오름차순으로 정렬(sort)
data2 = data_2019.set_index('연령대코드')
data2.sort_index()
```

```python
# groupby의 size() 함수를 사용해 연령대 별 주상병코드의 빈도 수를 추출
data3 = data2.groupby(by=['연령대코드', '주상병코드'])
data3.size()
```
