## 원하는 컬럼 뽑기

```python
# drop()
hospital_00 = hospital.drop(['인허가취소일자','재개업일자', '소재지면적', '완화의료지정형태', '지정취소일자', '완화의료담당부서명', '총인원', '구조사수', '최초지정일자'], axis=1)
hospital_00.info()
```

```python
# loc() 추천
hospital_01 = hospital_00.loc[:,['영업상태명','도로명전체주소','사업장명','좌표정보(x)','좌표정보(y)','의료기관종별명','진료과목내용','진료과목내용명']]
hospital_01.head()
```
