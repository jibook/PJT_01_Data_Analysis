```python
import pandas as pd
import numpy as np
```

### 데이터 불러오기

```python
df1 = pd.read_csv('./data/H_data/T20_2019_1백만.1.csv', encoding = 'cp949')
df1.head()
```

### 필요한 데이터만 추출

```python
var_list = ['연령대코드','성별코드', '주상병코드', '부상병코드', '입내원일수']
data = df1[var_list]
data.head()
```

```python
data.info()
```

### feature 데이터 만들기

#### 주상병코드 만들기

```python
var_list = ['연령대코드', '성별코드']
data_x = data[var_list]
data_x.head()
```

```python
# 가장 앞 알파벳만 따로 추출
alpha = []
for i in data['주상병코드']:
    alpha.append(i[0])
alpha
```

```python
# 숫자로 변환
alnum = []
for i in alpha:
    alnum.append(ord(i) - 64)
alnum
```

```python
# 컬럼 추가
data_x['주상병알파벳'] = alnum
data_x.head()
```

```python
# 나머지 숫자만 추출
num = []
for i in data['주상병코드']:
    num.append(i[1:])
num
```

```python
# 컬럼추가
data_x['주상병숫자'] = num
data_x.head()
```

```python
data_x.info()
```

```python
# 주상병숫자 타입변경
data_x['주상병숫자'] = data_x['주상병숫자'].str.replace('_','9999').astype(np.int64)
data_x.info()
```

#### 부상병코드 만들기

```python
# 가장 앞 알파벳만 따로 추출
alpha2 = []
for i in data['부상병코드']:
    alpha2.append(i[0])
alpha2
```

```python
# 숫자로 변환
alnum2 = []
for i in alpha2:
    alnum2.append(ord(i) - 64)
alnum2
```

```python
# 컬럼 추가
data_x['부상병알파벳'] = alnum2
data_x.head()
```

```python
# 나머지 숫자만 추출
num2 = []
for i in data['부상병코드']:
    num2.append(i[1:])
num2
```

```python
# 컬럼추가
data_x['부상병숫자'] = num2
data_x.head()
```

```python
data_x.info()
```

```python
# 부상병숫자 타입변경
data_x['부상병숫자'] = data_x['부상병숫자'].str.replace('_','9999').astype(np.int64)
data_x.info()
```

```python
data_x['입내원일수'] = data['입내원일수']
data_x.head()
```

### feature 데이터 확인

```python
data_x.head()
```

### train test 구분하기

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x.iloc[:,:6],
                                                    data_x.iloc[:,6:], test_size = 0.33)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, model.predict(X_test)))
```

```python
print(classification_report(y_test, model.predict(X_test)))
```
