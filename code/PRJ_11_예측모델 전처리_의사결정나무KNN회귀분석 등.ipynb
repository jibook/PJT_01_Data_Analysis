```python
import pandas as pd
import numpy as np
```

### 데이터 불러오기

```python
df1 = pd.read_csv('./data/H_data/T20_2019_1백만.1.csv', encoding = 'cp949')
df1.head()
```

```python
var_list = ['연령대코드','성별코드', '주상병코드', '부상병코드', '입내원일수']
data = df1[var_list]
data.info()
```

```python
df2 = pd.read_excel('./data/illness_ji.xlsx')
df2.head()
```

```python
df3 = pd.read_excel('./data/illness_jiji.xlsx')
df3.head()
```

### 질병코드에 해당하는 값(순위)으로 치환

```python
# 질병코드에 해당하는 값 치환하고 NaN은 사라짐

df2_ju = pd.merge(data, df2)
df2_ju
```

```python
# how 옵션으로 NaN값 남겨줌 
pd.merge(data, df2, how='outer',on= '주상병코드')


# 양쪽 컬럼이름이 다를때 합치는 법
df3_1 = pd.merge(data, df3, how='left', left_on='주상병코드', right_on='질병코드')
df3_1


# NaN 값 종류 확인하기
# set(data['주상병코드'].unique()) - set(df3['질병코드'].unique())
```

```python
print(df2_ju.head())
print(df3.head())
```

```python
data_ju = pd.merge(df2_ju, df3, how='left', left_on='주상병코드', right_on='질병코드')
data_ju
```

```python
varl = ['연령대코드','성별코드', '부상병코드', '입내원일수', '순위']
df2_ju_1 = df2_ju[varl]
df2_ju_1.head()
df2_ju_1
```

```python
data_23 = pd.merge(df2_ju_1, df3, how='right', left_on='부상병코드', right_on='질병코드')
data_23
```

```python
varll = ['연령대코드','성별코드', '입내원일수', '순위_x', '순위_y']
data_231 = data_23[varll]
data_231.head()
```

```python
data_231 = data_231.rename(columns={'순위_x':'주상병코드'})
data_231 = data_231.rename(columns={'순위_y':'부상병코드'})
```

```python
data_231
```

### train test 구분하기

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_231.iloc[:,:4],
                                                    data_231.iloc[:,4:], test_size = 0.33)
```

- 선형회귀 (소수점으로 나와서 실패)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept= True)
model
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

- 의사결정나무 (정확도가 너무 떨어져서 실패)

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, model.predict(X_test)))
```

```python
print(classification_report(y_test, model.predict(X_test)))
```
